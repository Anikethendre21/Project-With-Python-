# Import all libraries 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

# Import Files

```python
df_tracks=pd.read_csv('tracks.csv')
df_tracks.head()
```



# Data Cleansing Process

###  1. Cheking Null Values

```python
pd.isnull(df_tracks).sum()
```

### 2. Cheking Data Information

```python
df_tracks.info()
```

### 3. Least Top 10 artist's

```python
least_song = df_tracks.sort_values("popularity", ascending = True).head(10)
least_song
```

### 3.Discriptive statistics

```python
df_description = df_tracks.describe().transpose()
df_description
```

### 4. 10 Most Popular tracks which is gratter than 90

```python
most_popular = df_tracks.query("popularity > 90", inplace = False).sort_values("popularity", ascending = False)
most_popular[:10]
```

### 5. Set index to be release date Column.

```python
df_tracks.set_index("release_date", inplace = True)
pd.to_datetime(df_tracks.index)
df_tracks.head()
```

### 6. Checking Indexing

```python
df_tracks[["artists"]].iloc[18]
```

### 7. Duration Convert into ms to sec.

```python
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x:round(x/1000))
df_tracks.head()
```

### 8. Drop old Column.

```python
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

### 9.View Top 5 Listener

```python
top_listener=df_tracks.sort_values("duration", ascending = False)
top_listener.head(5)
```

# Visualization 

### Drop Unwanted Column for visualization.

corr_df = df_tracks.drop(["key","mode","explicit"], axis= 1 ).corr(method= "pearson")


### Insert Heatmap 

```python
corr_df = df_tracks.drop(["key","mode","explicit"], axis= 1 ).corr(method= "pearson")

plt.figure(figsize = (14,6))

heatmap =sns.heatmap(corr_df, annot= True, fmt=".1g", vmin=-1,vmax=1,center= 0, cmap="inferno",linewidths=1, linecolor= "Black")
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation= 90)
```

### Create 5% Data for regration plot

```python
short_data = df_tracks.sample(int(0.005*len(df_tracks)))
```

### Insert regration plot

```python
plt.figure(figsize=(10,6))
regplot = sns.regplot(data  =short_data , y = "loudness", x = "energy", scatter_kws ={"color": "#b93556"}, line_kws = {"color":"#fcffa4"})
regplot.set(title="Loudness VS Enerty Correlations")
```

```python
plt.figure(figsize=(10,6))
regplot = sns.regplot(data  =short_data , y = "popularity", x = "acousticness", scatter_kws ={"color": "#932667"}, line_kws = {"color":"Blue"})
regplot.set(title="Popularity VS Acousticness Correlations")
```

```python
df_tracks["dates"]= df_tracks.index.get_level_values("release_date")
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

### Insert Bar Plot For Songs Count Per Year S

```python
sns.displot(years,discrete = True,aspect=2,height=5,kind="hist",color = "#520e6d").set(title="Number of Song per Year")
```

```python
total_dr=df_tracks.duration
fig_dims= (18,7)
fig,ax= plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y= total_dr,ax = ax,errwidth = False).set(title = "Year VS Duration")
plt.xticks(rotation = 90)
```

```python
total_dr = df_tracks.duration
sns.set_style(style = "whitegrid")
fig_dims = (10,5)
fig, ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr,ax=ax, color = "#d44842").set(title="Year VS Duration")
plt.xticks(rotation=60)
```

```python
df_genre=pd.read_csv("SpotifyFeatures.csv")
df_genre.head()
```
