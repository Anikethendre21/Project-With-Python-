```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[40,30,50,60,68,90,42]
```

```python
plt.xlabel=("Day")
plt.ylabel=("Temp")
plt.title("Wether")

plt.plot(x,y,color="green",linewidth=1)
```

```python
plt.plot(x,y,"rD",linestyle="dotted",markersize=20)
```

```python
day=[1,2,3,4,5,6,7]
mini=[100,30,50,60,68,90,42]
maxi=[45,36,56,67,70,97,58]
avgi=[42,33,52,65,69,93,10]
```

```python
plt.xlabel=("Days")
plt.ylabel=("temp")
plt.title=("Wether")


plt.plot(day,maxi,label="Max")
plt.plot(day,mini,label="Min")
plt.plot(day,avgi,label="Avg")

plt.legend(loc="best",shadow=True,fontsize="small")
plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=["Google","Amazon","Facebook"]
y=[40,30,50]
p=[10,15,30]
```

```python
plt.bar(x,y)
```

```python
plt.xlabel('compnys')
plt.ylabel("Revenue(us$)")
plt.title("Ampaire")
plt.bar(x,y)
```

```python
x1=np.arange(len(x))
x1
```

```python
plt.xlabel('compnys')
plt.ylabel("Revenue(us$)")
plt.title("Ampaire")
plt.bar(x1-0.2,y,width=0.4,label="Revenue")
plt.bar(x1+0.2,p,width=0.4,label="Revenue")
plt.legend()
```

```python
plt.xlabel('compnys')
plt.ylabel("Revenue(us$)")
plt.title("Ampaire")
plt.barh(x1-0.2,y,label="Revenue")
plt.barh(x1+0.2,p,label="Revenue")
plt.legend()
```

```python
sugar=[113,58,90,150,89,88,93,115,95,80,77,82]
sugar1=[90,120,140,39,58,90,80,135,121,139,88,100]
```

```python
plt.hist(sugar,bins=[80,100,125,200],rwidth=0.95,color="y")
```

```python
plt.xlabel("Lavel")
plt.ylabel("Count of Pepoles")
plt.title("Suger Petient's")

plt.hist([sugar,sugar1],bins=[50,80,100,120,150,200],rwidth=0.95,color=["g","b"],
        label=["Men","Women"],orientation="horizontal")
plt.legend()
```
